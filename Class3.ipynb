```python
from groq import Groq

client = Groq()
completion = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
      {
        "role": "user",
        "content": "O que e IA"
      },
      {
        "role": "assistant",
        "content": "**Inteligência Artificial (IA)** é um ramo da ciência da computação que cria sistemas capazes de realizar tarefas que, tradicionalmente, exigiriam inteligência humana. Esses sistemas podem **perceber**, **raciocinar**, **aprender**, **tomar decisões** e **interagir** com o ambiente de forma autônoma ou semi‑autônoma.\n\n### Como funciona?\n- **Algoritmos**: Conjunto de regras matemáticas que processam dados.\n- **Dados**: Grandes volumes de informações (texto, imagens, áudio, sensores) que alimentam os algoritmos.\n- **Modelos**: Estruturas (ex.: redes neurais) que “aprendem” padrões a partir dos dados.\n- **Treinamento**: O modelo ajusta seus parâmetros para minimizar erros, usando técnicas como aprendizado supervisionado, não supervisionado ou por reforço.\n\n### Principais áreas da IA\n| Área | Descrição | Exemplos de aplicação |\n|------|-----------|-----------------------|\n| **Aprendizado de Máquina (Machine Learning)** | Algoritmos que aprendem a partir de dados. | Recomendação de produtos, detecção de fraudes. |\n| **Aprendizado Profundo (Deep Learning)** | Redes neurais com múltiplas camadas que capturam padrões complexos. | Reconhecimento de voz, visão computacional, geração de texto. |\n| **Processamento de Linguagem Natural (NLP)** | Compreensão e geração de linguagem humana. | Chatbots, tradutores automáticos, análise de sentimentos. |\n| **Visão Computacional** | Interpretação de imagens e vídeos. | Reconhecimento facial, veículos autônomos. |\n| **Robótica** | Integração de IA com hardware físico. | Robôs industriais, drones autônomos. |\n| **Sistemas Especialistas** | Regras codificadas para domínios específicos. | Diagnóstico médico, consultoria jurídica. |\n\n### Onde a IA está presente no dia a dia?\n- **Assistentes virtuais** (Siri, Alexa, Google Assistant)  \n- **Recomendações** em plataformas de streaming e e‑commerce (Netflix, Amazon)  \n- **Filtros de spam** e **detecção de fraudes** em bancos  \n- **Tradução automática** (Google Translate)  \n- **Carros autônomos** e **sistemas de assistência ao motorista** (Tesla Autopilot, piloto automático da Uber)  \n- **Ferramentas de produtividade** como corretores ortográficos inteligentes, geração de textos e códigos (ChatGPT, Copilot)\n\n### Limitações e desafios\n- **Viés nos dados**: se o conjunto de treinamento contém preconceitos, o modelo pode reproduzi‑los.  \n- **Interpretabilidade**: muitos modelos, especialmente redes neurais profundas, são “caixas‑pretas”, dificultando a compreensão de como chegaram a uma decisão.  \n- **Privacidade e segurança**: coleta massiva de dados pode comprometer informações pessoais.  \n- **Impacto no mercado de trabalho**: automação de tarefas pode substituir alguns empregos, exigindo requalificação profissional.\n\n### Futuro da IA\nA pesquisa avança em áreas como **IA explicável**, **IA geral (AGI)** – que busca capacidades cognitivas comparáveis às humanas – e **IA ética**, que procura garantir que os sistemas sejam justos, transparentes e responsáveis.\n\nEm resumo, a Inteligência Artificial é a disciplina que permite que máquinas imitem (e, em alguns casos, superem) habilidades humanas por meio de algoritmos que aprendem e se adaptam a partir de dados, transformando praticamente todos os setores da sociedade."
      },
      {
        "role": "user",
        "content": ""
      }
    ],
    temperature=1,
    max_completion_tokens=8192,
    top_p=1,
    reasoning_effort="medium",
    stream=True,
    stop=None
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```
