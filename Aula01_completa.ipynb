### Estruturas Condicionais

```python
nome = input("Digite o seu nome: ")
```

```python
idade = int(input("Digite a sua idade: "))
```

```python
type(idade)
```

```python
if idade >= 18:
    print(f"O {nome} pode entrar na festa porque ele é maior de idade.")
    print(f"Ele tem {idade} anos de idade.")
else:
    print(f"O {nome} NÃO pode entrar na festa porque ele é menor de idade.")
    print(f"Ele tem {idade} anos de idade.")
```

```python
media = float(input("Digite a sua média: "))

if media >= 7.0:
    print("Aprovado!")
elif media >= 5.0:
    print("Recuperação!")
else:
    print("Reprovado!")
```

```python
media = float(input("Digite a sua média: "))

if media < 5.0:
    print("Reprovado!")
elif media >= 5.0 and media < 7.0:
    print("Recuperação!")
else:
    print("Aprovado!")
```
