```python
dict_desafio = [
    {"nome": "Maria Silva", "media": 8.9},
    {"nome": "Joao dias", "media": 1.9},
    {"nome": "Ana Oliveira", "media": 2.5},
    {"nome": "Pedro bosta", "media": 4.6},
]
```

```python
n = 0

while n < len(dict_desafio):
  print(dict_desafio[n])
  n += 1
```

```python
for elemento in dict_desafio:
  print(elemento)
```

```python
for n in range(len(dict_desafio)):
  print(dict_desafio[n])
```

```python
pessoa = {
    "nome": "Fabricio",
    "idade": 19,
    "altura": 1.87,
}
```
